```python
matrix1 = [
    [10, 10, 1],   # Initials (JJA)
    [21, 21, 2]    # Second letters of names (Justine, Jude, Aborot)
]


matrix2 = [
    [2, 0, 2],     # First 3 digits of student number (202)
    [2, 6, 3]      # Last 3 digits of student number (263)
]


print("Matrix 1:")
for row in matrix1:
    print(row)

print("\nMatrix 2:")
for row in matrix2:
    print(row)


matrix3 = []
for i in range(2):
    row = []
    for j in range(3):
        row.append(matrix1[i][j] + matrix2[i][j])
    matrix3.append(row)

print("\nMatrix 3 (Addition of Matrix 1 and 2):")
for row in matrix3:
    print(row)

# 3. Scalar Multiplication
matrix4 = []
for i in range(2):
    row = []
    for j in range(3):
        row.append(matrix1[i][j] * 2)
    matrix4.append(row)

print("\nMatrix 4 (Matrix 1 multiplied by 2):")
for row in matrix4:
    print(row)

# 4. Matrix Transpose
matrix5 = []
for j in range(3):
    row = []
    for i in range(2):
        row.append(matrix2[i][j])
    matrix5.append(row)

print("\nMatrix 5 (Transpose of Matrix 2):")
for row in matrix5:
    print(row)

# 5. Matrix Multiplication
matrix6 = []
for i in range(2):
    row = []
    for j in range(2):
        total = 0
        for k in range(3):
            total += matrix3[i][k] * matrix5[k][j]
        row.append(total)
    matrix6.append(row)

print("\nMatrix 6 (Matrix 3 multiplied by Matrix 5):")
for row in matrix6:
    print(row)

# 6. Sum of All Elements in Matrix 3
sum_matrix3 = 0
for i in range(2):
    for j in range(3):
        sum_matrix3 += matrix3[i][j]

print("\nSum of all elements in Matrix 3:", sum_matrix3)

# 7. Zero Matrix
matrix7 = [
    [0, 0, 0],
    [0, 0, 0]
]

print("\nMatrix 7 (Zero Matrix 2x3):")
for row in matrix7:
    print(row)
```

```python
# Proof of clean data

# Removing any rows with missing values
df_cleaned = df.dropna()

# Checking for null values
null_check = df_cleaned.isnull().sum()

# Showing the number of rows after cleaning
row_count = df_cleaned.shape[0]

# Results
print("Null values in each column after cleaning:")
print(null_check)

print(f"\nTotal number of rows after cleaning: {row_count}")

# Proof condition
if row_count >= 100 and null_check.sum() == 0:
    print(" Data is clean and contains at least 100 rows.")
else:
    print(" Data is not clean or has fewer than 100 rows.")
```

```python
import numpy as np

prices = df['Price per Unit'].values
units_sold = df['Units Sold'].values
sales = df['Total Sales'].values
profits = df['Operating Profit'].values
```

```python
# Operations Using Numpy
# 1. Mean of Prices
mean_price = np.mean(prices)
print(f"1 Mean Price per Unit: {mean_price:.2f}")

# 2. Median of Units Sold
median_units = np.median(units_sold)
print(f"2 Median Units Sold: {median_units}")

# 3. Standard Deviation of Total Sales
std_sales = np.std(sales)
print(f"3 Standard Deviation of Total Sales: {std_sales:.2f}")

# 4. Maximum Profit
max_profit = np.max(profits)
print(f"4 Maximum Operating Profit: {max_profit:.2f}")

# 5. Minimum Profit
min_profit = np.min(profits)
print(f"5 Minimum Operating Profit: {min_profit:.2f}")

# 6. Total revenue (sum of Total Sales)
total_revenue = np.sum(df['Total Sales'].values)
print(f"6 Total Revenue (Sum of Sales): {total_revenue:.2f}")

# 7. Coefficient of Variation for Units Sold (standard deviation / mean)
units_std = np.std(df['Units Sold'].values)
units_mean = np.mean(df['Units Sold'].values)
coef_variation_units = units_std / units_mean
print(f"7 Coefficient of Variation (Units Sold): {coef_variation_units:.4f}")
```

```python
# Operation Using Scipy
from scipy.stats import kruskal
# Group data by Region
grouped_units = [group['Units Sold'].values for name, group in df.groupby('Region')]

# Performing Kruskal-Wallis H-test
h_stat, p_val = kruskal(*grouped_units)

print("= Kruskal-Wallis Test on Units Sold by Region =")
print(f"H-statistic: {h_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
if p_val < 0.05:
    print(" Statistically significant difference in Units Sold between regions.")
else:
    print(" No statistically significant difference in Units Sold between regions.")
```

```python
# Operation Using Statsmodel
import pandas as pd
import statsmodels.api as sm

# Cleaning the data by dropping any rows with missing values (Just in case)
df_cleaned = df.dropna()

# Defining independent variable
X = df_cleaned[['Price per Unit', 'Units Sold']] 

# Defining dependent variable
y = df_cleaned['Operating Profit'] 

# Adding constant for intercept
X = sm.add_constant(X)

# OLS regression
model = sm.OLS(y, X).fit()

# Display regression summary
print(model.summary())
```
